```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
#파일 불러오기
titanic  = pd.read_csv('/content/2025-1-BigData-Team/TitanicData/titanic1309.csv')
```

```
# 결측치 제거 (성별 및 생존 여부가 없는 행 제거)
titanic = titanic.dropna(subset=['Sex', 'Survived'])
```

```
# 파이 차트 그리기
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
```

```
#남성 생존률 파이 차트
male_counts = titanic[titanic['Sex'] == 'male']['Survived'].value_counts()
plt.pie(male_counts, labels=['Not Survived', 'Survived'], autopct='%1.1f%%', startangle=90, colors=['grey', 'blue'])
plt.title('Survival Rate - male')
plt.show()
```

```
#여성 생존률 파이 차트
female_counts = titanic[titanic['Sex'] == 'female']['Survived'].value_counts()
plt.pie(female_counts, labels=['Not Survived', 'Survived'], autopct='%1.1f%%', startangle=90, colors=['grey', 'red'])
plt.title('Survival Rate - Female')
plt.show()
```
